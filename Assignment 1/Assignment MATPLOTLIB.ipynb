```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open('Sample1.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x,y)
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Epic Information')
plt.legend()
plt.show()
```
