```python
X=data[:,:-1].copy()
```

```python
Y=data[:,-1].copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

```python
print(X_train.shape, Y_train.shape)
```

```python
Y_train=Y_train.reshape(-1,1)
```

```python
Y_test=Y_test.reshape(-1,1)
```

```python
print(X_test.shape, Y_test.shape)
```

```python
X_train=X_train.T
```

```python
X_test=X_test.T
```

```python
Y_train=Y_train.T
Y_test=Y_test.T
```

```python
def compute_cost(y,y_hat):
    return np.sqrt(np.sum(((y_hat-y)**2), axis = 1)/y.shape[1])
```

```python
# Method 1
def compute_w1(X,Y):
    w = Y.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T))
    return w
```

```python
def add_row_1(arr):
    ones = np.ones((1,arr.shape[1]))
    arr= np.concatenate((arr, ones), axis=0)
    return arr
```

```python
X_train_1=add_row_1(X_train)
print(X_train.shape)
```

```python
print(X_train)
```

```python
X_test_1 = add_row_1(X_test)
```

```python
print(X_test.shape)
print(X_test)
```

```python
print(X_test_1.shape,X_train_1.shape)
```

```python
w1=compute_w1(X_train_1, Y_train)
```

```python
print(w1)
```

```python
print(w1.shape)
```

```python
Y_hat_1 = w1.dot(X_train_1)
```

```python
cost1 = compute_cost(Y_train, Y_hat_1)
```
